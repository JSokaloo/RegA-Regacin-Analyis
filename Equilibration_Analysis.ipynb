# RMSD

```python
import numpy as np
import matplotlib.pyplot as plt
def window_average(x, N):
    return np.convolve(x, np.ones(N)/N, mode='valid')
```

```python
!./extract_energies.sh
```

```python
RMSD = np.loadtxt('rmsd.xvg', skiprows=18)
RG = np.loadtxt('gyrate.xvg', skiprows=27)
Energies = np.loadtxt('energy.xvg', skiprows=31)
```

```python
window_time = window_average(RMSD[:,0], 1000)
window_RMSD = window_average(RMSD[:,1], 1000)
```

```python
plt.plot(RMSD[:,0]/1000, RMSD[:,1]*10, c='g')
plt.plot(window_time/1000, window_RMSD*10, c='k')
plt.xlabel('Time (ns)')
plt.ylabel(r'RMSD ($\AA$)')
```

```python
window_time = window_average(RG[:,0], 1000)
window_Rg = window_average(RG[:,1], 1000)
```

```python
plt.plot(RG[:,0]/1000, RG[:,1]*10)
plt.plot(window_time/1000, window_Rg*10)
plt.xlabel('Time (ns)')
plt.ylabel(r'Rg ($\AA$)')
```
